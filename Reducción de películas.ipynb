```python
import pandas as pd
import glob

csv_files = glob.glob("*.csv")

dataframes = []
for file in csv_files:
    df = pd.read_csv(file)


C = df['rating'].mean()

m = df['votes'].quantile(0.9)

q_movies = df.copy().loc[df['votes'] >= m]

def weighted_rating(x, m=m, C=C):
    v = x['votes']
    R = x['rating']
    return (v / (v + m) * R) + (m / (m + v) * C)

q_movies['score'] = q_movies.apply(weighted_rating, axis=1)

q_movies = q_movies.sort_values('score', ascending=False)

print(q_movies[['movie_name', 'votes', 'rating', 'score', 'genre']].head(30))
```

```python
import pandas as pd
import glob

csv_files = glob.glob("*.csv")

recommendations_by_genre = []
for file in csv_files:
    df = pd.read_csv(file)

    C = df['rating'].mean()

    m = df['votes'].quantile(0.9)

    q_movies = df.copy().loc[df['votes'] >= m]

    # Definir la función weighted_rating
    def weighted_rating(x, m=m, C=C):
        v = x['votes']
        R = x['rating']
        # Cálculo basado en la fórmula de IMDB
        return (v / (v + m) * R) + (m / (m + v) * C)

    q_movies['score'] = q_movies.apply(weighted_rating, axis=1)

    q_movies = q_movies.sort_values('score', ascending=False)

    recommendations_by_genre.append(q_movies)

combined_df = pd.concat(recommendations_by_genre)

print(combined_df[['movie_name', 'votes', 'rating', 'score', 'genre']].head(30))
```

```python
import pandas as pd
import glob

csv_files = glob.glob("*.csv")

all_recommendations = pd.DataFrame()

for file in csv_files:
    df = pd.read_csv(file)

    df = df[['movie_id', 'movie_name', 'year', 'genre', 'rating', 'description', 'votes']]

    all_recommendations = pd.concat([all_recommendations, df])

all_recommendations.reset_index(drop=True, inplace=True)

all_recommendations.to_csv("recommendations.csv", index=False)

print("Todas las recomendaciones guardadas en el archivo 'recommendations.csv'")

```
